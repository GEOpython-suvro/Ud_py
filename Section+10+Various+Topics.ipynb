# 1- QUESTION:
Ask the user to enter a course code, like MATH 247 or CMSCI 125. The code consists of an alphabetic
department code followed by a space and then a course number. If the course number is not in the
range from 100 to 499, then output that the user’s entry is invalid.

```python
#Validate a course code in the form of some letters, a space, and then some numbers.  Numbers must be in range 100 to 499.
code = input('Enter course code: ')
num = int(code.split(' ')[1])
if num < 100 or num > 499:
    print('Invalid course code.')
else:
    print('Valid Course code')
```

# 2- QUESTION:
 Write a program that asks the user to enter several numbers separated by semicolons. Then print out the sum of all the numbers.

```python
#2 User enters numbers separated by semicolons, then add up numbers.
s = input('Enter some numbers separated by semicolons: ')
total = 0
for n in s.split(';'):
    total += int(n)
print(total)
```

# 3- QUESTION:
Write a program that use the examples below . Then write
a program that reads through the string and replaces each number n in the sentence with n + 

 
 1. For instance:
input: I bought 4 apples and 17 oranges.My favorite number is 8.
output: I bought 5 apples and 18 oranges. My favorite number is 9.

```python
#5 Replace all the numbers in a sentence with one more than their values.
s = 'I bought 4 apples and 17 oranges. input: My favorite number is 8.'
L = s.split()#['I', 'bought', '4', 'apples', 'and', '17', 'oranges.', 'input:', 'My', 'favorite', 'number', 'is', '8.']
M = []
for x in L:
    if x.isdigit():#checking if x integer or digit. if it is digit then append the code 
        M.append(str(int(x)+1))#iterate through each x string in splitted list, and find the number then add 1 to each
    elif x[:-1].isdigit():#there is punctiation rigth after each digit.so eliminate from the end
        M.append(str(int(x[:-1])+1) + x[-1]) # handle punctuation after a number
    elif x[1:].isdigit():#Puctuation elimination from the front of the number
        M.append(x[0] + str(int(x[1:])+1)) # handle punctuation before a number
    elif x[1:-1].isdigit():#eliminate punctuation both side
        M.append(x[0] + str(int(x[1:-1])+1) + x[-1]) # handle punctuation on both sides of a number
    else:
        M.append(x)
print(' '.join(M))
```

# 4- QUESTION:
Write a program that generates a list of 10 random multiplication problems, which are strings of a
form like 2 x 9 or 10x 7, where the two numbers can range from 2 to 12. At the same time, generate a
list of answers to those problems. Then loop through the list, asking the user to enter answers to the
problems. Print out whether each answer is right or wrong, and at the end print out how many the
user got correct.

```python
# Generate random multiplication problems and answers, then use them in a quiz.
from random import randint

Prob = []
Ans = []
for i in range(10):
    x = randint(2, 12)# random num is generated and asign to x variable
    y = randint(2, 12)# random num is generated and asign to y variable
    Prob.append(str(x) + ' x ' + str(y))#empty list is created and store the question of the multiplication
    Ans.append(x*y)#empty list is created and store the answer of multiplication questions
print(Prob)
print(Ans)
count = 0
for i in range(10):#There should be 10 iteration 
    guess = eval(input(Prob[i] + '? '))#Guess var holds the nums that users enter. Prob[i] is the value inside the prob list
    if guess == Ans[i]:#if guess is equal ans variable. ans[i] is the answer inside the list
        print('Right!')#print Right if the equality is True, and count it 
        count += 1
    else:
        print('Wrong.  Answer is', str(Ans[i]) + '.')#Otherwise, the answer is wrong. and show the answer.
print('You got', count, 'right.')#Show how many time is correct answer.

```

# 5- QUESTION:
Write a program that generates and prints a list of 20 random multiplication problems, each a string
like '2x 11', with the property that the random numbers being multiplied can range from 2 through
12, but can’t be 10. Do this in two ways:
    
(a) Use randint(2,12) to generate the random numbers and use a while loop to keep generating
until a number that is not 10 is generated.

(b) Use the choice() function on an appropriate list.

```python
# Generate 20 random multiplication problems with numbers from 2 to 12, but no 10s
#(a) Using randint and a while loop to avoid 10
from random import randint

L = []
for i in range(20):# list of 20 random multiplication problems
    x = randint(2, 12)# random numbers can  range from 2 through 12
    while x == 10:#x cant be equal 10 . So, we need to take a look at the stiuation that x ==10 and rewrite them
        x = randint(2, 12)# the code is rewriting the x while x ==10
    y = randint(2, 12) # y is running under for loop because y can be equal 10 while x not equal 10
    while y == 10:# this time we need to rewrite y value while y ==10
        y = randint(2, 12)#rewrite y
    L.append(str(x) + ' x ' + str(y)) # then append the values that outputs are different than 10
print(L)

#(b) Using choice
from random import choice
print(100*("*"))
P = [2,3,4,5,6,7,8,9,11,12]
L = [str(choice(P)) + ' x ' + str(choice(P)) for i in range(20)]
print(L)
```

# 6- QUESTION:
Write a program that asks the user for an angle in degrees-minutes-seconds form, such as 49d33'22'' or 153d4'22'', where the input will starts with the number of degrees (0 to 360), followed by the letter d, followed by the number of minutes (0 to 60), followed by an apostrophe, followed by the number of seconds (0 to 60), followed by two apostrophes.
The program should convert that angle to decimal form, rounded to two decimal places. For instance, 49d33'22 should become 49.56 and 153d4'22'' should become 153.07. To convert from
degrees/minutes/seconds to the decimal form, use d+m/60+s/3600 where d, m, and s are the numbers of degrees, minutes, and seconds, respectively.

```python
#Convert degrees-minutes-seconds form to decimal form
#input will starts with the number of degrees (0 to 360)
#followed by the letter d
# followed by the number of minutes (0 to 60)
#followed by an apostrophe
#followed by the number of seconds (0 to 60)
#followed by two apostrophes
s = input('Enter angle: ')
L = s.split('d')
degrees = int(L[0])
M = L[1].split("'")
minutes = int(M[0])
seconds = int(M[1])
print(degrees + minutes/60 + seconds/3600)
```

# 8- QUESTION:
Write a program that uses nested loops to print out all integer solutions to the equation 
37x+14y = 11
with x and y both between -100 and 100.

```python
#18 Use nested loops to find all integer solutions to 37x+14y=11 for x and y between -100 and 100.
for x in range(-100, 101):
    for y in range(-100, 101):
        if 37*x+14*y == 11:
            print(x, y)
```
